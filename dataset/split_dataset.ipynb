```python
import torch, random, os, numpy as np
from torch_geometric.data import Data 
```

```python
dataset = torch.load("/jbod/rokubo/ELNESDiff/dataset/Si_O_ELNES.pt")
```

```python
material_id_list = []
dataset = [data for data in dataset if data.x.shape[0] > 1]
print(len(dataset))
for data in dataset:
    material_id = data.id.split("_")[0]
    material_id_list.append(material_id)

material_id_list = list(set(material_id_list))
print(len(material_id_list))
```

```python
train_length = int(len(material_id_list) * 0.8)
valid_length = int(len(material_id_list) * 0.1)
test_length = len(material_id_list) - train_length - valid_length

torch.manual_seed(42)
random.shuffle(material_id_list, random.seed(42))
train_dataset_ids, valid_dataset_ids, test_dataset_ids = torch.utils.data.random_split(material_id_list, [train_length, valid_length, test_length])
```

```python
train_dataset = [data for data in dataset if data.id.split("_")[0] in train_dataset_ids]
valid_dataset = [data for data in dataset if data.id.split("_")[0] in valid_dataset_ids]
test_dataset = [data for data in dataset if data.id.split("_")[0] in test_dataset_ids]
print(f"train dataset_length: {len(train_dataset)}")
print(f"valid dataset_length: {len(valid_dataset)}")
print(f"test dataset_length: {len(test_dataset)}")
print(f"total dataset_length: {len(train_dataset) + len(valid_dataset) + len(test_dataset)}")
```

```python
torch.save(train_dataset, "/jbod/rokubo/ELNESDiff/dataset/Si_O_ELNES_train.pt")
torch.save(valid_dataset, "/jbod/rokubo/ELNESDiff/dataset/Si_O_ELNES_valid.pt")
torch.save(test_dataset, "/jbod/rokubo/ELNESDiff/dataset/Si_O_ELNES_test.pt")
```
